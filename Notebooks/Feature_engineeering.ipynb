```python
df = pd.read_csv("cleaned_Data.csv")
df.head()
```

```python

binary_cols = [
    'Is Debtor',
    'Tuition Fees Up-to-Date',
    'Special Educational Needs',
    'Scholarship Holder'
]

df[binary_cols] = df[binary_cols].apply(lambda col: col.map({"Yes":1, "No":0}) if col.dtype=='object' else col)
```

```python
df.head()
```

```python
df['Pass Rate (1st Sem)'] = df.apply(
    lambda row: row['Approved Units (1st Sem)'] / row['Enrolled Units (1st Sem)'] 
    if row['Enrolled Units (1st Sem)'] != 0 else 0, axis=1
)
df['Pass Rate (2nd Sem)'] = df.apply(
    lambda row: row['Approved Units (2nd Sem)'] / row['Enrolled Units (2nd Sem)'] 
    if row['Enrolled Units (2nd Sem)'] != 0 else 0, axis=1
)
```

```python
df['Total Approved Units'] = df['Approved Units (1st Sem)'] + df['Approved Units (2nd Sem)']
df['Total Enrolled Units'] = df['Enrolled Units (1st Sem)'] + df['Enrolled Units (2nd Sem)']
df['Total Evaluated Units'] = df['Evaluated Units (1st Sem)'] + df['Evaluated Units (2nd Sem)']
```

```python

df['Grade Difference'] = df['Average Grade (2nd Sem)'] - df['Average Grade (1st Sem)']
```

```python
def load_category(x):
    if x >= 40:
        return 'Full Load'
    elif x >= 20:
        return 'Medium Load'
    else:
        return 'Low Load'

df['Load Category'] = df['Total Enrolled Units'].apply(load_category)
```

```python

df['Debt Risk'] = df['Is Debtor'].apply(lambda x: 1 if str(x).strip().lower() == 'yes' else 0)
```

```python
df['Tuition Stable'] = df['Tuition Fees Up-to-Date'].apply(lambda x: 1 if str(x).strip().lower() == 'yes' else 0)
```

```python
df['Socioeconomic Stress Score'] = (df['Unemployment Rate (%)'] + df['Inflation Rate (%)'])
```

```python

df['Financial Hardship'] = (df['Inflation Rate (%)'] / df['GDP per Capita (USD)'])
```

```python

df['Econ_Disadvantage'] = (
    (df['Is Debtor']) +
    (1 - df['Tuition Fees Up-to-Date']) +
    (df['Special Educational Needs']) +
    (df['Scholarship Holder'] == 1)
)
```

```python
df.to_csv("cleaned_feature_eng_data.csv", index = False)
```
