```python

df['Evaluated Units (1st Sem)'] = df['Evaluated Units (1st Sem)'].clip(upper=df['Enrolled Units (1st Sem)'])
df['Evaluated Units (2nd Sem)'] = df['Evaluated Units (2nd Sem)'].clip(upper=df['Enrolled Units (2nd Sem)'])
```

```python
# 1st Semester
# Rule 1: If Evaluated = 0 and Not Evaluated = 0 → Evaluated = Enrolled
mask_rule1_1st = (
    (df['Evaluated Units (1st Sem)'] == 0) &
    (df['Not Evaluated Units (1st Sem)'] == 0)
)
df.loc[mask_rule1_1st, 'Evaluated Units (1st Sem)'] = df.loc[mask_rule1_1st, 'Enrolled Units (1st Sem)']

# Rule 2: If Not Evaluated != 0 → Evaluated = Enrolled - Not Evaluated
mask_rule2_1st = (df['Not Evaluated Units (1st Sem)'] != 0)
df.loc[mask_rule2_1st, 'Evaluated Units (1st Sem)'] = (
    df.loc[mask_rule2_1st, 'Enrolled Units (1st Sem)'] - 
    df.loc[mask_rule2_1st, 'Not Evaluated Units (1st Sem)']
)

# 2nd Semester
# Rule 1
mask_rule1_2nd = (
    (df['Evaluated Units (2nd Sem)'] == 0) &
    (df['Not Evaluated Units (2nd Sem)'] == 0)
)
df.loc[mask_rule1_2nd, 'Evaluated Units (2nd Sem)'] = df.loc[mask_rule1_2nd, 'Enrolled Units (2nd Sem)']

# Rule 2
mask_rule2_2nd = (df['Not Evaluated Units (2nd Sem)'] != 0)
df.loc[mask_rule2_2nd, 'Evaluated Units (2nd Sem)'] = (
    df.loc[mask_rule2_2nd, 'Enrolled Units (2nd Sem)'] - 
    df.loc[mask_rule2_2nd, 'Not Evaluated Units (2nd Sem)']
)
```

```python

mask_fix_ne_1st = (
    (df['Not Evaluated Units (1st Sem)'] == 0) &
    (df['Enrolled Units (1st Sem)'] != df['Evaluated Units (1st Sem)'])
)

df.loc[mask_fix_ne_1st, 'Not Evaluated Units (1st Sem)'] = (
    df.loc[mask_fix_ne_1st, 'Enrolled Units (1st Sem)'] -
    df.loc[mask_fix_ne_1st, 'Evaluated Units (1st Sem)']
)
```

```python

# 1st Semester
inconsistent_1st = df[
    df['Not Evaluated Units (1st Sem)'] != 
    (df['Enrolled Units (1st Sem)'] - df['Evaluated Units (1st Sem)'])
]

# 2nd Semester
inconsistent_2nd = df[
    df['Not Evaluated Units (2nd Sem)'] != 
    (df['Enrolled Units (2nd Sem)'] - df['Evaluated Units (2nd Sem)'])
]

print("1st Sem inconsistent rows:", len(inconsistent_1st))
print("2nd Sem inconsistent rows:", len(inconsistent_2nd))
```

```python
numerical_cols = [
    'Age at Enrollment','Unemployment Rate (%)', 'Inflation Rate (%)', 'GDP per Capita (USD)'
]
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Cap values outside the range
    df[col] = df[col].clip(lower, upper)
```
